```python
nlp=spacy.load('en_core_web_sm')
```

since when we upload the file it is stored as key and value key is file name and value is the content so for extracting the keys will give the file name 0 is for extracting the first value


```python
SKILLS_DB = [
    "python", "sql", "machine learning", "deep learning", "nlp",
    "data science", "pandas", "numpy", "tensorflow", "pytorch",
    "git", "excel", "power bi", "java", "c++"
]

def extract_skills(text):
  doc=nlp(text.lower())
  tokens=[token.lemma_ for token in doc if not token.is_stop]
  text_pp=" ".join(tokens)
  found_skills=[]
  for skill in SKILLS_DB:
    if skill in text_pp:
      found_skills.append(skill)
  return found_skills

```

```python
def skill_gap_analyzer(canditate_text):
  job_description = """
  Looking for a candidate with Python, SQL, Machine Learning,
  NLP and Git experience.
  """
  job_skills=set(extract_skills(job_description))
  canditate_text=set(extract_skills(canditate_text))
  matched_skills=job_skills & canditate_text
  missing_skills =job_skills-canditate_text
  match_percentage = (
        len(matched_skills) / len(job_skills) * 100
        if job_skills else 0
    )

  status = "ELIGIBLE ‚úÖ" if match_percentage >= 70 else "NOT ELIGIBLE ‚ùå"

  return (
    "üìå Job Skill Eligibility Report\n\n"
    f"Required Skills:\n- " + "\n- ".join(sorted(job_skills)) + "\n\n"
    f"Your Skills:\n- " + "\n- ".join(sorted(canditate_text)) + "\n\n"
    f"Matched Skills:\n- " + "\n- ".join(sorted(matched_skills)) + "\n\n"
    f"Missing Skills:\n- " + (
        "\n- ".join(sorted(missing_skills)) if missing_skills else "None üéØ"
    ) + "\n\n"
    f"Match Percentage: {match_percentage:.2f}%\n"
    f"Final Decision: {status}"
)

```
